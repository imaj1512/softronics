points from news articles while ensuring factual accuracy and readability.

```python
from transformers import pipeline  # Import pipeline for easy model usage
from datasets import load_dataset  # Import load_dataset for loading datasets
```

```python
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")  # Load a pre-trained BART model
```

```python
article_text = """
Your news article text goes here.
"""

summary = summarizer(article_text, max_length=130, min_length=30, do_sample=False)  # Generate summary
print(summary[0]["summary_text"])  # Print the summary text
```
